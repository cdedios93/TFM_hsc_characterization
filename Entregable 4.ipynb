## Visualization: the function

Taking this into consideration, to ease the visualization and analysis of the excluded HSC, we decided to create a function to analyze and show each channel from each HSC. Please find attached the function below:

```python
# Importing libraries
import cv2
import matplotlib.pyplot as plt
# Being imgroute the folder were the images are; z, xpos and ypos the coordinates of the cell; and surrpix the pixels to create the rectangle of the images (considering xpos and ypos = 0)
def cell_analyzer(imgroute, z, xpos, ypos,surrpix):
    # Image load
    if z < 10:
        red = cv2.imread(imgroute + "/T00001C01Z00" + str(z+1) + ".tif",0) 
        white = cv2.imread(imgroute + "/T00001C02Z00" + str(z+1) + ".tif",0) 
        blue = cv2.imread(imgroute + "/T00001C03Z00" + str(z+1) + ".tif",0)
    else:
        red = cv2.imread(imgroute + "/T00001C01Z0" + str(z+1) + ".tif",0) 
        white = cv2.imread(imgroute + "/T00001C02Z0" + str(z+1) + ".tif",0) 
        blue = cv2.imread(imgroute + "/T00001C03Z0" + str(z+1) + ".tif",0)
    # Channel plotting
    plt.figure(figsize = (20,20))
    plt.subplot(1,3,1),plt.imshow(red[(ypos-surrpix):(ypos+surrpix),(xpos-surrpix):(xpos+surrpix)],cmap="hot")
    plt.title("Red channel")
    plt.subplot(1,3,2),plt.imshow(white[(ypos-surrpix):(ypos+surrpix),(xpos-surrpix):(xpos+surrpix)],cmap="gray")
    plt.title("White channel")
    plt.subplot(1,3,3),plt.imshow(blue[(ypos-surrpix):(ypos+surrpix),(xpos-surrpix):(xpos+surrpix)],cmap="Blues")
    plt.title("Blue channel")
```
